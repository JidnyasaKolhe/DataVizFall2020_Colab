```
df_covid = df_covid_confirmed
df_covid.head()
```

```
## Get all countries:
Countries = df_covid['Country/Region'].unique()
Countries
```

```
## Get Spain, Germany and France data
df_covid_c = df_covid[df_covid['Country/Region'].isin(['Spain','Germany','Poland','Belgium'])].copy()                            
df_covid_c.head(20)
```

```
## keep the ones which Province/State == NaN
df_covid_c = df_covid_c[df_covid_c['Province/State'].isnull()]
df_covid_c.T
```

```
df_covid_c = df_covid_c.T.iloc[4:]
df_covid_c.columns=['Belgium','Germany','Poland','Spain']
df_covid_c.tail(10)
```

```
df_covid_c=df_covid_c.diff().fillna(0)
```

```
df_covid_c=df_covid_c.rolling(window=14).mean().fillna(0)
```

```
## let us check the column types
df_covid_c.dtypes
```

```
## let us convert Date to pandas datetime object
df_covid_c.index = pd.to_datetime(df_covid_c.index)
```
